```python
from bs4 import BeautifulSoup # this module helps in web scrapping.
import requests  # this module helps us to download a web page
```

```python
import pandas as pd
```

```python
url ='https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)'
```

```python
# get the contents of the webpage in text format and store in a variable called data
data  = requests.get(url).text
```

```python
soup = BeautifulSoup(data,"html.parser")
```

```python
#find all html tables in the web page
tables = soup.find_all('table') # in html table is represented by the tag <table>
```

```python
len(tables)
```

```python
for index,table in enumerate(tables):
    if ("GDP (US$ million) by country" in str(table)):
        table_index = index
print(table_index)
```

```python
print(tables[table_index].prettify())
```

```python
gdp_data = pd.DataFrame(columns=[ "Country", "Region", "Estimate", "Year"])

for row in tables[table_index].tbody.find_all("tr"):
    col = row.find_all("td")
    if (col != []):
        Country = col[1].text
        Region = col[2].text.strip()
        Estimate = col[3].text.strip()
        Year = col[4].text.strip()
        gdp_data = gdp_data.append({ "Country":Country, "Region":Region, "Estimate":Estimate, "Year":Year}, ignore_index=True)

gdp_data
```
